```python
# question 3
import numpy as np
def addTwoNumbers(l1, l2):
    
    sum_l1=0
    sum_l2=0
    for i in range(len(l1)):
        sum_l1 +=l1[i]*np.power(10,len(l1)-1-i)     
    for i in range(len(l1)):
        sum_l2 +=l2[i]*np.power(10,len(l2)-1-i)    
    sum=sum_l1+sum_l2 
    listNode=[]
    while sum:      
        listNode.append(sum%10)
        sum=sum//10    
    return listNode
    
        
    
l1=[2,4,3,1]
l2=[5,6,4,5]
addTwoNumbers(l1, l2)
```

```python
# question 4
'''Given a string s, find the length of the longest substring without repeating characters. 
You can expect the string length is less than 100, and only contains English letters.
Example 1:
Input: s = "abcabcbb"
Output: 3

'''
def lengthOfLongestSubstring(string):
    
    seen=[]
    old=[]
    for i in (list(string)):
        if i not in seen:
            seen.append(i)
        else:
            if len(seen)>len(old):
                old=seen.copy()
            seen=[]
            seen.append(i)
           
    return max(len(seen),len(old))
    
string='abcabcbefgd'
lengthOfLongestSubstring(string)
```

```python
#question 8
'''Use SciPy for an application of Discrete Fourier Transform (DFT), modify the question8.py, 
and apply DFT to the array “a” and visualize both original and result signals'''

from matplotlib import pyplot as plt
import numpy as np 
import scipy

#Frequency in terms of Hertz
fre  = 5 
#Sample rate
fre_samp = 50
t = np.linspace(0, 2, 2 * fre_samp, endpoint = False )
a = np.sin(fre  * 2 * np.pi * t)
figure, axis = plt.subplots()
axis.plot(t, a)
axis.set_xlabel ('Time (s)')
axis.set_ylabel ('Signal amplitude')
plt.show()


T=1/fre
N=fre_samp
yf =scipy.fft.fft(a)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
figure, axis = plt.subplots()
axis.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
axis.set_xlabel ('Frequency(Hz)')
axis.set_ylabel ('Amplitude')
axis.grid()
plt.show()
```

```python
# extra question4
'''
Use template matching with OpenCV to find Messi’s face in an image, try all 6 methods and plot the result.
Modify the extra_question5/code.py
'''
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('messi5.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

face = cv2.imread('messi_face.jpg')
face=cv2.cvtColor(face,cv2.COLOR_BGR2RGB)

methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

for m in methods:
        
    image_copy=img.copy()
    
    method=eval(m)
         
    res=cv2.matchTemplate(image_copy,face,method)   # 比较相似度
    
    min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        
        top_left=min_loc
    else:
        
        top_left=max_loc
    
    height,width,channels=face.shape
    
    bottom_right=(top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(image_copy,top_left,bottom_right,(255,0,0),10)
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title('HEATMAP')
    
    plt.subplot(122)
    plt.imshow(image_copy)
    plt.title('{:s}'.format(m))
    plt.show()
```
